```python
def display(board):
    for i in range(0,len(board)):
        if not (i+1)%3 == 0:
            print(f"{board[i]}",end="   |")
        elif i+1==3 or i+1==6:
            print(f"{board[i]}",end="")
            print("\n")
            print("____|____|____",end="")
            print("\n")
        elif i+1==9:
            print(board[i])

def accept(board):
    pos_check=False
    while not pos_check:
        pos=input("Enter the number to check :\n")
        if pos.isdigit():
            pos_check=True
            pos=int(pos)
            if not pos in range(1,10) and pos_check:
                pos_check=False
                print("Number out of range(1-9)")
            elif pos_check and not board[pos-1]=='':
                pos_check=False 
                print("Position not vacant")
        else:
            print("Please enter in digits")
    return pos-1

def check(board,s):
    val = (board[0] == s and board[1] == s and board[2] == s) or (board[3] == s and board[4] == s and board[5] == s) or (board[6] == s and board[7] == s and board[8] == s) or (board[0] == s and board[3] == s and board[6] == s) or (board[1] == s and board[4] == s and board[7] == s) or (board[2] == s and board[5] == s and board[8] == s) or (board[0] == s and board[4] == s and board[8] == s) or (board[2] == s and board[4] == s and board[6] == s)
    return not val

def main():
    symbols=['X','O']
    player1_symbol=input("Enter player 1 symbol : \n").upper()
    symbols.remove(player1_symbol)
    player2_symbol=symbols[0]
    print(f"Player1 Symbol = {player1_symbol}")
    print(f"Player2 Symbol = {player2_symbol}")
    running=True
    board=['','','','','','','','','']
    display(board)
    while running:
        print("Player 1 Input(1-9) :")
        p1_pos=accept(board)
        board[p1_pos]=player1_symbol
        display(board)
        running=check(board,player1_symbol)
        if running==False:
            print("Player 1 wins!")
            break
        print("Player 2 Input(1-9)")
        p2_pos=accept(board)
        board[p2_pos]=player2_symbol
        display(board)
        running=check(board,player2_symbol)
        if running==False:
            print("Player 2 wins!")
            break

main()
play = True
while play:
    p = input("Want to play another game?(Y/N)\n")
    if p == 'Y' or p == 'y':
        main()
    elif p == 'N' or p == 'n':
        play = False
```
