# Currency Converter

Create a currency converter with a graphical user interface (GUI) that allows users to convert between different currencies using real-time exchange rates. Users can enter an amount, select the source currency and the target currency, and get the converted amount.

````python
# Currency Converter: A script that converts currencies from one to another,
# using up-to-date exchange rates.

# tkinter: It facilities  Graphical User interfaces (GUIs).
# pip install tkinter

# forex_python:  It is a Free Foreign exchange rate and currency conversion library.
# pip install forex_python

#pip install tkinter
import tkinter as tk 
from tkinter import *
import tkinter.messagebox
#GUI
root = tk.Tk()
 
root.title("Currency converter")
 
Tops = Frame(root, bg = '#e6e5e5', pady=2, width=1850, height=100, relief="ridge")
Tops.grid(row=0, column=0)
 
headlabel = tk.Label(Tops, font=('lato black', 19, 'bold'), text='Currency converter ',
                    bg='#e6e5e5', fg='black')
headlabel.grid(row=1, column=0, sticky=W)
 
variable1 = tk.StringVar(root)
variable2 = tk.StringVar(root)
 
variable1.set("currency")
variable2.set("currency")
#Function To For Real Time Currency Conversion
 
def RealTimeCurrencyConversion():
    from forex_python.converter import CurrencyRates
    c = CurrencyRates()
 
    from_currency = variable1.get()
    to_currency = variable2.get()
 
    if (Amount1_field.get() == ""):
        tkinter.messagebox.showinfo("Error !!", "Amount Not Entered.\n Please a valid amount.")
 
    elif (from_currency == "currency" or to_currency == "currency"):
        tkinter.messagebox.showinfo("Error !!",
                                    "Currency Not Selected.\n Please select FROM and TO Currency form menu.")
 
    else:
        new_amt = c.convert(from_currency, to_currency, float(Amount1_field.get()))
        new_amount = float("{:.4f}".format(new_amt))
        Amount2_field.insert(0, str(new_amount))
 
#clearing all the data entered by the user
def clear_all():
    Amount1_field.delete(0, tk.END)
    Amount2_field.delete(0, tk.END)
 
 
CurrenyCode_list = ["INR", "USD", "CAD", "CNY", "DKK", "EUR"]
 
root.configure(background='#e6e5e5')
root.geometry("700x400")
 
Label_1 = Label(root, font=('lato black', 27, 'bold'), text="", padx=2, pady=2, bg="#e6e5e5", fg="black")
Label_1.grid(row=1, column=0, sticky=W)
 
label1 = tk.Label(root, font=('lato black', 15, 'bold'), text="\t    Amount  :  ", bg="#e6e5e5", fg="black")
label1.grid(row=2, column=0, sticky=W)
 
label1 = tk.Label(root, font=('lato black', 15, 'bold'), text="\t    From Currency  :  ", bg="#e6e5e5", fg="black")
label1.grid(row=3, column=0, sticky=W)
 
label1 = tk.Label(root, font=('lato black', 15, 'bold'), text="\t    To Currency  :  ", bg="#e6e5e5", fg="black")
label1.grid(row=4, column=0, sticky=W)
 
label1 = tk.Label(root, font=('lato black', 15, 'bold'), text="\t    Converted Amount  :  ", bg="#e6e5e5", fg="black")
label1.grid(row=8, column=0, sticky=W)
 
Label_1 = Label(root, font=('lato black', 7, 'bold'), text="", padx=2, pady=2, bg="#e6e5e5", fg="black")
Label_1.grid(row=5, column=0, sticky=W)
 
Label_1 = Label(root, font=('lato black', 7, 'bold'), text="", padx=2, pady=2, bg="#e6e5e5", fg="black")
Label_1.grid(row=7, column=0, sticky=W)
 
FromCurrency_option = tk.OptionMenu(root, variable1, *CurrenyCode_list)
ToCurrency_option = tk.OptionMenu(root, variable2, *CurrenyCode_list)
 
FromCurrency_option.grid(row=3, column=0, ipadx=45, sticky=E)
ToCurrency_option.grid(row=4, column=0, ipadx=45, sticky=E)
 
Amount1_field = tk.Entry(root)
Amount1_field.grid(row=2, column=0, ipadx=28, sticky=E)
 
Amount2_field = tk.Entry(root)
Amount2_field.grid(row=8, column=0, ipadx=31, sticky=E)
 
Label_9 = Button(root, font=('arial', 15, 'bold'), text="   Convert  ", padx=2, pady=2, bg="lightblue", fg="white",
                 command=RealTimeCurrencyConversion)
Label_9.grid(row=6, column=0)
 
Label_1 = Label(root, font=('lato black', 7, 'bold'), text="", padx=2, pady=2, bg="#e6e5e5", fg="black")
Label_1.grid(row=9, column=0, sticky=W)
 
Label_9 = Button(root, font=('arial', 15, 'bold'), text="   Clear All  ", padx=2, pady=2, bg="lightblue", fg="white",
                 command=clear_all)
Label_9.grid(row=10, column=0)
 
 
root.mainloop()
````

````python
import tkinter as tk
from tkinter import messagebox
from forex_python.converter import CurrencyRates

def real_time_currency_conversion():
    try:
        amount = float(amount_entry.get())
        from_currency = from_currency_var.get()
        to_currency = to_currency_var.get()

        if from_currency == "Select Currency" or to_currency == "Select Currency":
            messagebox.showerror("Error", "Please select currencies.")
            return

        c = CurrencyRates()
        converted_amount = c.convert(from_currency, to_currency, amount)

        result_label.config(text=f"Converted Amount: {converted_amount:.2f} {to_currency}")
    except ValueError:
        messagebox.showerror("Error", "Invalid amount entered.")

def clear_all():
    amount_entry.delete(0, tk.END)
    result_label.config(text="Converted Amount:")

# GUI setup
root = tk.Tk()
root.title("Currency Converter")

curreny_codes = ["USD", "EUR", "GBP", "JPY", "AUD"]
from_currency_var = tk.StringVar(root)
from_currency_var.set("Select Currency")
to_currency_var = tk.StringVar(root)
to_currency_var.set("Select Currency")

amount_label = tk.Label(root, text="Amount:")
amount_label.pack()

amount_entry = tk.Entry(root)
amount_entry.pack()

from_currency_option = tk.OptionMenu(root, from_currency_var, *curreny_codes)
from_currency_option.pack()

to_currency_option = tk.OptionMenu(root, to_currency_var, *curreny_codes)
to_currency_option.pack()

convert_button = tk.Button(root, text="Convert", command=real_time_currency_conversion)
convert_button.pack()

result_label = tk.Label(root, text="Converted Amount:")
result_label.pack()

clear_button = tk.Button(root, text="Clear All", command=clear_all)
clear_button.pack()

root.mainloop()
````

````python
# First, sign up for a free account on the Open Exchange Rates website to obtain an API access key.

# Import the necessary libraries:

import requests
import json

# Define the function to convert the currency:

def convert_currency(amount, from_currency, to_currency, api_key):
    # Define the API endpoint and parameters
    url = f"https://openexchangerates.org/api/convert/{amount}/{from_currency}/{to_currency}"
    payload = {
        "app_id": api_key,
    }
    
    # Send the API request and get the response
    response = requests.get(url, params=payload)
    if response.ok:
        # Extract the converted amount from the response
        response_data = json.loads(response.content.decode('utf-8'))
        return response_data['result']
    else:
        # Return an error message if the API request failed
        return "Error: Unable to convert currency"

#Call the function with the amount, from currency, to currency, and API access key:

converted_amount = convert_currency(amount=100, from_currency="USD", to_currency="EUR", 
                                    api_key="your-open-exchange-rates-api-key-here")
print(converted_amount)


# This is just a basic structure, and you can customize it further based on your requirements and the 
# API you choose to use. Please make sure to use this script ethically and not for any malicious purposes. 
# Additionally, note that some APIs may have rate limits or other restrictions on usage, so be sure to
# review the API documentation for any such restrictions before implementing your script.
````
