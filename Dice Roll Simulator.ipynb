# Dice Roll Simulator
___

Create a dice roll simulator with a graphical user interface (GUI) that allows users to simulate rolling dice. Users can specify the number of dice and the number of sides on each dice. The simulator displays the resulting rolls in a visually engaging manner.

```python
# Dice Roll Simulator: A script that simulates the rolling of dice, allowing the
# user to specify the number of dice and sides.

# Importing tkinter and random modules
from tkinter import *
import random
  
# Creating a window
window = Tk()
  
# Setting background colour of the window
window.configure(bg="black")
  
# Setting geometry of the window
window.geometry("550x350")
  
# Setting title of the window
window.title("Rolling The Dices Game")
  
# Preventing maximizing of the window
window.resizable(0, 0)
  
# Creating function to roll the dices
def roll_dices():
     # These values indicate dots on the dices.
     # For eg: \u2680 corresponds to 1 dot,
     # \u2681 corresponds to 2 dots etc.
    dice_dots = ['\u2680', '\u2681',
                 '\u2682', '\u2683',
                 '\u2684', '\u2685']
    # Generating random dots on the dices
    label.configure(
        text=f'{random.choice(dice_dots)}{random.choice(dice_dots)}')
    label.pack()
  
  
# Adding button to roll the dices
roll_button = Button(window, text="Roll!",
                     width=10, height=2,
                     font=15, bg="aqua",
                     bd=2, command=roll_dices)
# Setting the position of the button
roll_button.pack(padx=10, pady=15)  
  
# Adding Label
label = Label(window, font=("times", 250),
              bg="black", fg="yellow")
  
window.mainloop()
```

```python
# Import the necessary libraries:

import tkinter as tk
import random

# Create the tkinter window:

window = tk.Tk()
window.title("Dice Roll Simulator")
window.geometry("400x300")

# Create the labels and entry widgets for the number of dice and sides:

num_dice_label = tk.Label(window, text="Number of dice:")
num_dice_label.pack()

num_dice_entry = tk.Entry(window)
num_dice_entry.pack()

num_sides_label = tk.Label(window, text="Number of sides:")
num_sides_label.pack()

num_sides_entry = tk.Entry(window)
num_sides_entry.pack()

# Create the function to roll the dice:

def roll_dice():
    num_dice = int(num_dice_entry.get())
    num_sides = int(num_sides_entry.get())
    
    results = []
    for i in range(num_dice):
        result = random.randint(1, num_sides)
        results.append(result)
    
    result_label.config(text=f"Results: {results}")

# Create the button to roll the dice:

roll_button = tk.Button(window, text="Roll Dice", command=roll_dice)
roll_button.pack()

# Create the label to display the results:

result_label = tk.Label(window, text="Results:")
result_label.pack()

# Run the tkinter event loop:

window.mainloop()
```
